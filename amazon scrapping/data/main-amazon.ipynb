```python
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium import webdriver
from pyspark.sql import SparkSession
import time
import random

# Initialize Spark Session
spark = SparkSession.builder.appName("SeleniumParallel").getOrCreate()
sc = spark.sparkContext

# Function to scrape a single page
def scrape_page(page_num):
    chrome_options = webdriver.ChromeOptions()
    chrome_options.add_argument("--headless")  # Run in headless mode
    chrome_options.add_argument("--no-sandbox")
    chrome_options.add_argument("--disable-dev-shm-usage")
    chrome_options.add_argument(
        "user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
    )

    driver = webdriver.Chrome(options=chrome_options)
    wait = WebDriverWait(driver, 10)

    data_list = []

    if page_num == 1:
        url = "https://www.amazon.in/s?i=computers&rh=n%3A1375424031&s=popularity-rank&fs=true&ref=lp_1375424031_sar"
    else:
        url = f"https://www.amazon.in/s?i=computers&rh=n%3A1375424031&s=popularity-rank&fs=true&qid=1739175275&page={page_num}"

    # Retry mechanism for loading the page
    attempts = 0
    while attempts < 5:
        try:
            driver.get(url)
            wait.until(EC.presence_of_element_located((By.XPATH, "//*[@role='listitem']")))
            print(f"Page {page_num} loaded successfully!")
            break

        except Exception:
            sleep_time = random.randint(3, 12)
            print(f"Page {page_num} failed to load. Retrying {attempts + 1} time(s)... Sleeping for {sleep_time} sec")
            time.sleep(sleep_time)

            old_tab = driver.current_window_handle
            driver.execute_script("window.open('', '_blank');")  
            driver.switch_to.window(driver.window_handles[-1])  
            driver.get(url)  
            time.sleep(3)  
            driver.switch_to.window(old_tab)
            driver.close()
            driver.switch_to.window(driver.window_handles[0])  
            attempts += 1

    if attempts == 5:
        print(f"Skipping page {page_num} after multiple failed attempts.")
        driver.quit()
        return []

    try:
        driver.find_element(By.TAG_NAME, "body").send_keys("\ue00f")  # PAGE_DOWN
        time.sleep(2)

        items = driver.find_elements(By.XPATH, "//*[@role='listitem']")
        print(f"Page {page_num}: Found {len(items)} items.")

        for item in items:
            try:
                name = item.find_element(By.XPATH, ".//h2").text
                link = item.find_element(By.CLASS_NAME, "a-link-normal").get_attribute("href")
                ratings_elements = item.find_elements(By.CSS_SELECTOR, ".a-size-base.s-underline-text")
                ratings = ratings_elements[0].text if ratings_elements else "No Ratings"
                total_buys = item.find_element(By.CSS_SELECTOR, ".a-size-base.a-color-secondary").text
                price = (
                    item.find_element(By.CLASS_NAME, "a-price-symbol").text
                    + " "
                    + item.find_element(By.CLASS_NAME, "a-price-whole").text
                )

                try:
                    deal_available = item.find_element(By.CLASS_NAME, "a-badge-text").text
                except:
                    deal_available = "No Deal"

                try:
                    original_price = item.find_element(
                        By.XPATH, '//span[contains(@class, "a-text-price") and contains(@class,"a-price")]/span[contains(@class, "a-offscreen")]'
                    ).get_attribute("textContent")
                except:
                    original_price = "Not Available"

                try:
                    coupon_savings = item.find_element(
                        By.XPATH, '//span[contains(@class,"s-coupon-unclipped")]/span[contains(@class,"s-coupon-highlight-color")]'
                    ).text
                except:
                    coupon_savings = "No Coupon"

            
                data_list.append(
                    {
                        "Name": name,
                        "Link": link,
                        "Ratings": ratings,
                        "Total Buys": total_buys,
                        "Price": price,
                        "Deal Available": deal_available,
                        "Original Price": original_price,
                        "Coupon Savings": coupon_savings,
                    }
                )

            except Exception as e:
                print(f"Error extracting data on page {page_num}: {e}")
    except Exception as e:
        print(f"Error retrieving items on page {page_num}: {e}")

    print(f"Total items scraped from page {page_num}: {len(data_list)}")
    driver.quit()
    return data_list

# List of pages to scrape
page_numbers = list(range(1, 16))

# Parallelize scraping across Databricks worker nodes
rdd = sc.parallelize(page_numbers, numSlices=5)  # Adjust numSlices based on Databricks cluster size
scraped_rdd = rdd.flatMap(scrape_page)  # Distribute scraping

# Collect results
final_data = scraped_rdd.collect()

# # Convert to Spark DataFrame for analysis
# df = spark.createDataFrame(final_data)
```

```python
schema = StructType([
    StructField("Name", StringType(), True),
    StructField("Link", StringType(), True),
    StructField("Ratings", StringType(), True),
    StructField("Total Buys", StringType(), True),
    StructField("Price", StringType(), True),
    StructField("Deal Available", StringType(), True),
    StructField("Original Price", StringType(), True),
    StructField("Coupon Savings", StringType(), True)
])
df = spark.createDataFrame(final_data, schema=schema)
```
